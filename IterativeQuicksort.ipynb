```python
#Quicksort without recursion
#Cassie Lease (2/8/26)
def Partition(A,p,r):
    x = A[r]
    i = p - 1
    for j in range(p,r):
        if ( A[j] <= x ):
            i = i + 1
            #swap A[i] <--> A[j]
            temp=A[i]
            A[i]=A[j]
            A[j]=temp
    #swap A[i+1] <--> A[r]
    temp=A[i+1]
    A[i+1]=A[r]
    A[r]=temp

    return i + 1
```

```python
#https://stackoverflow.com/questions/12553238/quicksort-iterative-or-recursive#comment16906885_12553238
#https://www.geeksforgeeks.org/python/python-program-for-iterative-quick-sort/
#Sources used

def Quicksort(A, l, h):
    size = h - l + 1
    stack = [0] * size
    top = -1

    top += 1
    stack[top] = l
    top += 1
    stack[top] = h

    while top >= 0:
        h = stack[top]
        top -= 1
        l = stack[top]
        top -= 1

        p = Partition(A, l, h)
        print(A)

        if p - 1 > l:
            top += 1
            stack[top] = l
            top += 1
            stack[top] = p - 1

        if p + 1 < h:
            top += 1
            stack[top] = p + 1
            top += 1
            stack[top] = h
```

```python
#Basic test of quicksort with comments
A = list(range(8))
random.shuffle(A)
print("Initial A: ", A)
Quicksort(A, 0, len(A)-1)
print("Final A: ", A)
```
